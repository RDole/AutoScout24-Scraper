```python
def getAllLinks(link):
    soup = getSoup(link)
    zs = [[soup.findAll('dl')[x].findAll('dt', {'class':'sc-ellipsis'})] for x in range(0,2)]
    tds = zs.findAll('dt', {'class':'sc-ellipsis'})
    return [td for td in tds]

links = getAllLinks('https://www.autoscout24.de/angebote/volkswagen-tiguan-1-4-tsi-bluemotion-technology-trend-fun-benzin-weiss-73302319-253f-4781-8b0a-849c475d065d?cldtidx=1')
links
```

```python
def getAllLinks(link):
    soup = getSoup(link)
    for x in range(0,2):
        zs= soup.findAll('dl')[x]
        for i in range(0,1):
            tds=zs.findAll('dt', {'class':'sc-ellipsis'})[i]
    return [td for td in tds]

links = getAllLinks('https://www.autoscout24.de/angebote/volkswagen-tiguan-1-4-tsi-bluemotion-technology-trend-fun-benzin-weiss-73302319-253f-4781-8b0a-849c475d065d?cldtidx=1')
links
```
