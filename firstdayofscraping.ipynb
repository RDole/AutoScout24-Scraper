```python
def getSoup(link):
    r = requests.get(link)
    r.encoding = 'UTF-8'
    return BeautifulSoup(r.text,'lxml')
```

```python
def getTitle(link):
    soup = getSoup(link)
    return soup.find('h2').text
```

```python
def getPrice(link):
    soup = getSoup(link)
    return soup.find('div', {"class":"cldt-price"}).find('h2').text[1:-1]
```
